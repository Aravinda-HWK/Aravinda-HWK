```python
from datasets import load_dataset

import pandas as pd

raw_datasets = load_dataset("haturusinghe/sinhala_off-sinhala-to-english")

dataframe = pd.DataFrame(raw_datasets["train"])

dataframe.head()
```

```python
!pip install transformers
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
checkpoint = "meta-llama/Llama-2-7b-hf"
model = AutoModelForCausalLM.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
def removeInput(output,input):
  # Find the index where the substring starts
  start_index = output.find(input)

  # Check if the substring is present in the original string
  if start_index != -1:
    # Remove the substring by slicing the original string
    modified_string = output[:start_index] + output[start_index + len(input):]
    return modified_string
  else:
    return ""

# Loop through the first five rows
for index, row in dataframe.head().iterrows():
    # Extract the instruction
    instruction = row["instruction"]

    # Extract input text
    input_text = row["input"]

    # Tokenize the prompt
    input_ids = tokenizer.encode(str(instruction)+str(input_text), return_tensors="pt")

    # Generate text using the model
    outputs = model.generate(input_ids,max_length=150)

    # Decode the generated text
    decoded_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the answer (might require further processing)
    # answer = decoded_text.split(": ")[-1].strip()
    print(f"Input: {input_text}")
    print(f"Answer: {removeInput(decoded_text,str(instruction)+str(input_text))}")
```
